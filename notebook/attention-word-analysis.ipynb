```python
def word2sen(list_of_word):
    sentences = []
    for sent in list_of_word:
        sentences.append(' '.join(sent))
    return sentences
```

```python
dct = {"a": 1, "b": 2, "c": 3, "d": 4, "e": 5}
k = 3
print(sorted(dct.keys(), reverse=True)[:k])
print(sorted(dct.items(), reverse=True)[:k])
```

We choose `sorted` to get the key and value. 

```python
from heapq import nlargest

def find_max_word(list_of_dict, k=3):
    # k means how many words we want to return
    attention_list = []
    for sent in list_of_dict: # sent is dictionary
        attention_list.append(sorted(sent.items(), reverse=True)[:k])
    return attention_list
```

```python
train_att_words = find_max_word(train_att_weights_list, k=3)
test_att_words = find_max_word(test_att_weights_list, k=3)
```
